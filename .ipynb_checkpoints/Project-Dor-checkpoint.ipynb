# Pre Processing

### Data Imputaion & Outliers Removal

##### is_news feature

There are no '0' in the data. We assume that all '?' are '0'.

We will also convert the string "1" to an int 1.

```python
# If value is '?' replace with 0, else convert to integer - for the train set
train.is_news = train.is_news.apply(lambda x: 0 if x == '?' else int(x))
# Same for the test set
test.is_news = test.is_news.apply(lambda x: 0 if x == '?' else int(x))
```

##### news_front_page feature

There are 16.8% of missing values ('?').

About 80% of the records are 0 (not a news front page).

We will impute the missing values with 0.

```python
# Display the the frequancy of each value
print (train.news_front_page.value_counts() / train.news_front_page.count())
```

```python
# If value is '?' replace with 0, else convert to integer - for the train set
train.news_front_page = train.news_front_page.apply(lambda x: 0 if x == '?' else int(x))
# Same for the test set
test.news_front_page = test.news_front_page.apply(lambda x: 0 if x == '?' else int(x))
```

##### alchemy_category feature

Replace missing values with 'unknown' category

```python
train.alchemy_category = train.alchemy_category.apply(lambda x: 'unknown' if x == '?' else x)
test.alchemy_category = test.alchemy_category.apply(lambda x: 'unknown' if x == '?' else x)
```

##### alchemy_category_score feature

Impute the missing values with the mean and transform the values from strings to floats

```python
# Transforme the strings to floats
train.alchemy_category_score = train.alchemy_category_score.apply(lambda x: float(x) if x != '?' else x)
test.alchemy_category_score = test.alchemy_category_score.apply(lambda x: float(x) if x != '?' else x)
```

```python
# Calcule the mean of the exsisting values
mean = train.alchemy_category_score[train.alchemy_category_score != '?'].mean()

# Impute the missing values with the train's mean (both the train and test sets)
train.alchemy_category_score = train.alchemy_category_score.apply(lambda x: mean if x == '?' else x)
test.alchemy_category_score = test.alchemy_category_score.apply(lambda x: mean if x == '?' else x)
```

### Outliers Removal

We'll remove outliers by removing records that are lower than the 1% precentile **or** higher than the 99% precentile in **any** of the features

```python
# lower and upper bounds
l, u = 0.01, 0.99

# Set a NaN value where the value is lower than the 1% precentile or higher than the 99% precentile
train_new = train.copy()
train_new[numeric_features] = train_new[numeric_features][(train_new[numeric_features] <= train_new[numeric_features].quantile(u)) & (train_new[numeric_features] >= train_new[numeric_features].quantile(l))]

# Remove a record if it contains at least 1 NaN value
train_new.dropna(how='any', inplace=True)

print ('Number of records in the train set after outliers removal: {}'.format(train_new.count()[0]))
```

### ROC & AUC Calculation

A function for plotting the ROC Curve

```python
def plot_ROC(fpr, tpr, auc):
    plt.figure(figsize=(10,7))
    plt.plot(fpr, tpr,lw=2, label='ROC Curve (AUC = {:.2f})'.format(auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0, 1])
    plt.ylim([0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
    
def plot_multiple_ROC(models_res):
    fig=plt.figure(figsize=(10,7), num=1)
    rocs = []
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

    for model in models_res:
        plt.plot(model['fpr'], model['tpr'],lw=2, label='{} (AUC = {:.2f})'.format(model['name'], model['auc']), figure=fig)
        plt.xlim([0, 1])
        plt.ylim([0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

### Cross Validation

# Initial Models

### KNN

We will use KNN model as a basic initial model

##### We've created a knn_classifier function:

**Input:**

k, train data, train labes, whether we're using k-fold cv, whether we want to plot, if we trying to predict the test (pred=True), test data (fetures).

**Output:**
- If in train mode (with or without k-fold),the function returns fpr, tpr, AUC (of the last set if with k-fold), the average AUC and accuracy, and plots ROC if specified.

- If in test mode returns the predection for the test set.

```python
def knn_classifier(k, knn_features, label, k_fold=False, k_fold_idx=None, plot=False, pred=False, test_knn=None):
    accuracy = []
    auc = []
    
    if k_fold:
        # Set the train and test sets according to the K-Fold indices (the loop will run 5 times)
        for train_idx, test_idx in k_fold_idx:
            train_features = knn_features.iloc[train_idx].copy()
            train_label = label.iloc[train_idx].copy()
            test_features = knn_features.iloc[test_idx].copy()
            test_label = label.iloc[test_idx].copy()

            # Create a KNN calssifier object
            knn_class = KNeighborsClassifier(n_neighbors=k)
            knn_class.fit(train_features, train_label)

            # Calculate the accuracy
            accuracy.append(knn_class.score(test_features, test_label))

            # Get the probabilities (confidance of the classification) - for calculating AUC
            probs = knn_class.predict_proba(test_features)[:,1]

            # Calculate the ROC and AUC
            fpr, tpr, thresholds = metrics.roc_curve(test_label, probs, pos_label=1)
            auc.append(metrics.auc(fpr,tpr))


        # Calculate and print the average accuracy and AUC values (for each k, average of each set in the K-Fold)    
        avg_accuracy = sum(accuracy) / float(len(accuracy))
        avg_auc = sum(auc) / float(len(auc))

        if plot:
            plot_ROC(fpr, tpr, auc[-1])
        
        return fpr, tpr, avg_auc, avg_accuracy
    
    # If not k_fold
    else: 
        # Create a KNN calssifier object
        knn_class = KNeighborsClassifier(n_neighbors=k)
        knn_class.fit(knn_features, label)
        
        # If not predicting the test
        if not pred:
            # Calculate the accuracy
            accuracy.append(knn_class.score(knn_features, label))

            # Get the probabilities (confidance of the classification) - for calculating AUC
            probs = knn_class.predict_proba(knn_features)[:,1]

            # Calculate the ROC and AUC
            fpr, tpr, thresholds = metrics.roc_curve(label, probs, pos_label=1)
            auc.append(metrics.auc(fpr,tpr))

            if plot:
                # Plot one of the ROC curves
                plot_ROC(fpr, tpr, auc[0])
            return fpr, tpr, auc[0], accuracy[0]
        
        # If predicting the test
        else:
            pred = knn_class.predict_proba(test_knn)[:,1]
            return pred
```

### Run The KNN Model

The chosen features for the KNN model are the numeric features **excluding** the features we infered that are less important in the feature importance and the PCA analysis (embed_ratio ,avglinksize, compression_ratio, image_ratio, spelling_errors_ratio).

It is possible to add binary features too as the data is normalized to the range [0,1], but we will not add them because of our conclusions from the feature selection stage.

The feature selection stage showed that the categorial feature **'alchemy_category'** seems to be important and explains the varince well (both from the feature importance and the PCA analysis). In order to use it in this model we will transform it into 12 binary features (one for each category).

We'll try differnt values for k and choose the one which will result with best average AUC value.

Note that the ROC curve is of the last sets in the k-fold sets

```python
# Selected features for KNN
features_for_knn = ['alchemy_category_score', 'commonlinkratio_3', 'frameTagRatio', 'html_ratio', 'linkwordscore',
 'non_markup_alphanum_characters', 'numberOfLinks', 'numwords_in_url', 'parametrizedLinkRatio']

# Set the train data for the KNN model (only numeric features)
knn_features = train_new[features_for_knn].copy()
# Add the categorial feature 'alchemy_category' as 12 binary features (one for each category)
knn_features = pd.concat([pd.get_dummies(train_new.alchemy_category), knn_features], axis=1)

# Set a list of k values to check
ks = range(3,59,2)


knn_best = {}
knn_auc = []
knn_accuracy = []

# Check the knn model for each k
for k in ks:
    fpr, tpr, auc, accuracy = knn_classifier(k ,knn_features, train_new.label, k_fold=True, k_fold_idx=k_fold_indices)
    knn_auc.append(auc)
    knn_accuracy.append(accuracy)
    
    # Update the best AUC model
    if auc >= knn_best.get('auc', 0):
        knn_best['auc'], knn_best['k'], knn_best['fpr'], knn_best['tpr'], knn_best['accuracy'] = auc, k, fpr, tpr, accuracy

# Plot the AUC and accuracy for each k
plt.figure(figsize=(10,7))
auc_plot = plt.scatter(ks, knn_auc, s=50, c='navy')
accuracy_plot = plt.scatter(ks, knn_accuracy, s=50,c='r')
plt.title('Average AUC & Accuracy - KNN')
plt.ylabel('Average Value')
plt.xlabel('K (Number of Neighbors)')
plt.xticks(ks)
plt.xlim([ks[0]-2,ks[-1]+2])
plt.legend([auc_plot, accuracy_plot], ['Average AUC', 'Average Accuracy'], loc='upper right')
plt.show()

# The best k
print ('\nThe Best K:')
print ('The k which resulted with highest average AUC is k={}'.format(knn_best['k']))
print ('The average AUC for k={} is: {:.3f}'.format(knn_best['k'], knn_best['auc']))
print ('The average accuracy for k={} is: {:.2f}% \n'.format(knn_best['k'], knn_best['accuracy']*100))
plot_ROC(knn_best['fpr'], knn_best['tpr'], knn_best['auc'])

# Save the the ROC and AUC for later (to compare with other models)
knn_roc_auc = knn_best
knn_roc_auc['name'] = 'knn'
```

```python
# Choose the features for the random forest model as explained above
features_for_rand_forest = list(train_new.columns.values.copy())
features_for_rand_forest.remove('label') # remove the label
features_for_rand_forest.remove('url')
features_for_rand_forest.remove('boilerplate')
```

We will first tune the random forest model hyper parameters using exhaustive grid search with different values for the following parameters:
- The minimum number of samples in order to split
- The trees maximum depth - it is important to set it as we are using a large number of features (regulation method)
- The criterion - whther to use entropy or gini index

The number of trees is chosen to be 50 due to time complexity reasons. After we will find the best parameters we will use a larger number of trees.

We are setting the random seed in order to avoid random influence.

```python
# Set the train data for the random forest model
rand_forest_features = train_new[features_for_rand_forest].copy()

# Parmeters grid to be checked
tuned_parameters = {'min_samples_split':[4, 8, 16, 20], 'max_depth':[5,8,10,20,30],'criterion':['gini', 'entropy'],
                   'n_estimators':[50], 'random_state':[1]}


forest_classifier = GridSearchCV(RandomForestClassifier(), tuned_parameters, cv=5,
                   scoring='roc_auc')
forest_classifier.fit(rand_forest_features, train_new.label)

params = forest_classifier.best_params_

print ('Best parameters are: {}'.format(params))
```

### Run the random forest model

With the parameters found above

```python
accuracy = []
auc = []

# Set the train and test sets according to the K-Fold indices (the loop will run 5 times)
for train_idx, test_idx in k_fold_indices:
    train_features = rand_forest_features.iloc[train_idx].copy()
    train_label = train_new.label.iloc[train_idx].copy()
    test_features = rand_forest_features.iloc[test_idx].copy()
    test_label = train_new.label.iloc[test_idx].copy()
    
    # Create a random forest classifier (with the relevant parameters) and train it
    forest = RandomForestClassifier(max_depth = params['max_depth'], min_samples_split=params['min_samples_split'],
                                    n_estimators=1000, criterion=params['criterion'],
                                    random_state = 1)
    forest = forest.fit(train_features, train_label)
    
    # Get the probabilities (confidance of the classification) - for calculating AUC
    probs = forest.predict_proba(test_features)[:,1]
    
    # Calculte ROC and AUC
    fpr, tpr, thresholds = metrics.roc_curve(test_label, probs, pos_label=1)
    auc.append(metrics.auc(fpr,tpr))
    
    # Calculte the accuracy
    accuracy.append(forest.score(test_features, test_label))

    
avg_accuracy = sum(accuracy) / float(len(accuracy))
avg_auc = sum(auc) / float(len(auc))
print ('The average AUC is: {:.3f}'.format(avg_auc))
print ('The average accuracy is: {:.2f}%'.format(avg_accuracy*100))

# Plot one of the ROC curves
plot_ROC(fpr, tpr, auc[-1])

# Save the the ROC and AUC for later (to compare with other models)
forest_roc_auc = {'name':'forest', 'fpr':fpr, 'tpr':tpr, 'auc':avg_auc, 'accuracy':avg_accuracy}
```

# Feature Engineering

### Bag of Words

Create a function to read the text from the json, remove non-letters and convert to lower case.

The text is extracted from the title, body and url.

```python
def extract_text(boiler):
    js = json.loads(boiler)
    if ('body' in js and js['body'] != None):
        body = js['body']
        # Remove non-letters        
        letters_only = re.sub("[^a-zA-Z]", " ", body) 
        # Convert to lower case, split into individual words
        words = letters_only.lower().split()
        body_text = ( " ".join(words))
    else:
        body_text = ""
        
    if ('title' in js and js['title'] != None):
        title = js['title']
        # Remove non-letters        
        letters_only = re.sub("[^a-zA-Z]", " ", title) 
        # Convert to lower case, split into individual words
        words = letters_only.lower().split()
        title_text = ( " ".join(words))
    else:
        title_text = ""
        
    if ('url' in js and js['url'] != None):
        url = js['url']
        # Remove non-letters        
        letters_only = re.sub("[^a-zA-Z]", " ", url) 
        # Convert to lower case, split into individual words
        words = letters_only.lower().split()
        url_text = ( " ".join(words))
    else:
        url_text = ""
    
    return body_text + title_text + url_text
```

##### Extract the text form the train and the test sets

(using the function above)

```python
train_text = train_original.boilerplate.apply(lambda x: extract_text(x))
test_text = test_original.boilerplate.apply(lambda x: extract_text(x))
```

#### Create new features using Bag of Words

We will use bag of words to create 5000 new features - the 5000 most popular words. For each record the value will be the frequancy of the word in the record's text.

We will remove stop words.

```python
# Create bag of words object and train it
vectorizer = CountVectorizer(analyzer = "word", tokenizer = None, preprocessor = None,
                             stop_words = {'english'}, max_features = 5000) 

train_data_features = vectorizer.fit_transform(train_text)

# Create a data frame for the new train features
train_data_features = train_data_features.toarray()
train_bow = pd.DataFrame(train_data_features)#, columns=[vectorizer.get_feature_names()])
# Add the urlid and set it as an index
train_bow['urlid'] = train_original.index.values
train_bow.set_index('urlid', inplace=True)

# Transform the test and save the new test features to a data frame
test_data_features = vectorizer.transform(test_text)
test_data_features = test_data_features.toarray()
test_bow = pd.DataFrame(test_data_features)#, columns=[vectorizer.get_feature_names()])
# Add the urlid and set it as an index
test_bow['urlid'] = test_original.index.values
test_bow.set_index('urlid', inplace=True)
```

### Random Forest with Bag of Words

We will use a Random forest model using (only) the new 5000 features

```python
accuracy = []
auc = []

# Create a random forest classifier (with the relevant parameters)
forest_bow = RandomForestClassifier(max_depth=20, min_samples_split=8,
                                n_estimators=1000, criterion='entropy',
                                random_state = 1)

# Set the train and test sets according to the K-Fold indices (the loop will run 5 times)
for train_idx, test_idx in k_fold_indices:
    train_features = train_bow.iloc[train_idx].copy()
    train_label = train_original.label.iloc[train_idx].copy()
    test_features = train_bow.iloc[test_idx].copy()
    test_label = train_original.label.iloc[test_idx].copy()
    

    forest_bow = forest_bow.fit(train_features, train_label)
    
    # Get the probabilities (confidance of the classification) - for calculating AUC
    probs = forest_bow.predict_proba(test_features)[:,1]
    
    # Calculte ROC and AUC
    fpr, tpr, thresholds = metrics.roc_curve(test_label, probs, pos_label=1)
    auc.append(metrics.auc(fpr,tpr))
    
    # Calculte the accuracy
    accuracy.append(forest_bow.score(test_features, test_label))

    
avg_accuracy = sum(accuracy) / float(len(accuracy))
avg_auc = sum(auc) / float(len(auc))
print ('The average AUC is: {:.3f}'.format(avg_auc))
print ('The average accuracy is: {:.2f}%'.format(avg_accuracy*100))

# Plot one of the ROC curves
plot_ROC(fpr, tpr, auc[-1])

# Save the the ROC and AUC for later (to compare with other models)
forest_bow_roc_auc = {'name':'forest_bow', 'fpr':fpr, 'tpr':tpr, 'auc':avg_auc, 'accuracy':avg_accuracy}
```
